```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)
y1 = 2*x -2
y2 = -0.5*x + 1/2
y3 = -x + 4

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.axis([-10,10,-10,10])
plt.show()
```

System of equations is equivalent to:

 image.png




This system is overconstrained. This means that **b** is not in the column space of **A**. But we know our least squares solution:

image.png

```python
A = np.matrix([ [2,-1], [1,2], [1,1] ])
print (A)
b = np.matrix([ [2], [1], [4] ])
print(b)

xstar = (A.T * A).I * A.T * b
print(xstar[0])
x_coords = xstar.flatten()
print(x_coords)
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.scatter(xstar[0, 0], xstar[1, 0], marker='x', s=100, color='black', label="Least Squares Solution")
plt.axis([-10,10,-10,10])
plt.show()
```
